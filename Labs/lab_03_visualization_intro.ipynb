## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```

- 1.1 Create a pie chart showing the count of cars by transmission type.  Display labels showing the percentage of each transmission type.  
- 1.2 Create a bar chart of cars and HP, with different colors for each car model
- 1.3 Create a scatter plot of wt (weight) vs. mpg.
- 1.4 Create a bar chart showing the average mpg for each cyl.
- 1.5 Plot a histogram of the weight column.

```python
import matplotlib.pyplot as plt

transmission_counts = df['am'].value_counts()

# Map numerical transmission type to labels for better understanding
transmission_labels = {0: 'Automatic', 1: 'Manual'}
transmission_counts.index = transmission_counts.index.map(transmission_labels)

plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Transmission Types')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a list of colors, one for each car model
# Using a colormap to get a distinct color for each car
colors = plt.cm.jet(np.arange(len(df.index)) / float(len(df.index)))

plt.figure(figsize=(12, 6))
plt.bar(df['carmodel'], df['hp'], color=colors)
plt.xlabel('Car Model')
plt.ylabel('Horsepower (hp)')
plt.title('Horsepower by Car Model')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Scatter plot of wt (weight) vs. mpg
df.plot.scatter(x='wt', y='mpg', title='Weight vs. MPG')
```

```python
# Bar chart showing the average mpg for each cyl
average_mpg_by_cyl = df.groupby('cyl')['mpg'].mean()

plt.figure(figsize=(8, 6))
average_mpg_by_cyl.plot(kind='bar', title='Average MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Histogram of the weight column
df['wt'].plot(kind='hist', bins=10, title='Weight Distribution')
```
