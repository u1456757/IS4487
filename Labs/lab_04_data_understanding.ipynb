## Part 3: Basic Exploration

Use `groupby` and `value_counts` to summarize trends across neighborhoods and cities.  The concept of data exploration will be covered more in week 5.


```python
# Average price by neighborhood
df.groupby('nhood')['price'].mean().sort_values(ascending=False).head(10)
```

```python
# Top cities by count
df['city'].value_counts().head(10)
```

### üîß Try It Yourself ‚Äì Part 3

Seek to understand the data by performing both of the following:

1. Group the listings by `year` and calculate the average price for each year.
2. Use `.value_counts()` on the `room_in_apt` column to see how common room rentals are.

### In Your Response:

1. Add a short comment or markdown cell describing any trends or insights you found.


```python
# Add code here üîß
print("Average price by year:")
print(df.groupby('year')['price'].mean().sort_values(ascending=False))

print("\nRoom in apartment counts:")
print(df['room_in_apt'].value_counts())
```

Average Price by Year: Rental prices generally increased over the years, peaking around 2015-2018, with 2015 showing the highest average price and 2000 the lowest.
Room in Apartment Counts: The vast majority of listings are for entire apartments, with only a very small number representing single rooms within an apartment.
