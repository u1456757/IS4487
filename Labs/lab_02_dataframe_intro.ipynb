## Getting data into Pandas

In this case we will load data from the statsmodels.org library

MTCARS is a traditional dataset used for learning.



```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
df['carmodel'] = df.index
print(df)
```


### ðŸ”§ Try It Yourself: Task 1 - Work with DataFrame

ðŸ”§ Assignment Tasks - Practice selecting specific rows and columns.
- 1.1 Extract the mpg (miles per gallon) column and calculate its mean.
- 1.2 Create a new dataframe containing only the rows where mpg is greater than 25.
- 1.3 Create a new dataframe with rows with the number of cylinders (cyl) is 4.  Include only the columns mpg and hp.



```python
# 1.1 Extract the mpg (miles per gallon) column and calculate its mean.
mpg_column = df['mpg']
mpg_mean = mpg_column.mean()
print(f"Mean of mpg: {mpg_mean}")

# 1.2 Create a new dataframe containing only the rows where mpg is greater than 25.
df_mpg_gt_25 = df[df['mpg'] > 25]
print("\nDataFrame with mpg > 25:")
print(df_mpg_gt_25.head())

# 1.3 Create a new dataframe with rows where the number of cylinders (cyl) is 4. Include only the columns mpg and hp.
df_cyl4_mpg_hp = df[df['cyl'] == 4][['mpg', 'hp']]
print("\nDataFrame with 4 cylinders, mpg and hp columns:")
print(df_cyl4_mpg_hp.head())
```


### ðŸ”§ Try It Yourself - Task 2 - Introduce derived columns.
- 2.1 Create a new column power_to_weight_ratio as hp / wt (horsepower divided by weight).

```python
# 2.1 Create a new column power_to_weight_ratio as hp / wt (horsepower divided by weight).
df['power_to_weight_ratio'] = df['hp'] / df['wt']
print("DataFrame with new 'power_to_weight_ratio' column (first 5 rows):")
print(df[['hp', 'wt', 'power_to_weight_ratio']].head())
```

### ðŸ”§ Try It Yourself - Task 3 - Summarize data.
- 3.1 Group the data by the number of cylinders (cyl) and calculate the average mpg for each group.
- 3.2 Find the maximum horsepower (hp) for cars in each gear group.

```python
# 3.1 Group the data by the number of cylinders (cyl) and calculate the average mpg for each group.
avg_mpg_by_cyl = df.groupby('cyl')['mpg'].mean()
print("\nAverage mpg by number of cylinders:")
print(avg_mpg_by_cyl)

# 3.2 Find the maximum horsepower (hp) for cars in each gear group.
max_hp_by_gear = df.groupby('gear')['hp'].max()
print("\nMaximum horsepower by gear group:")
print(max_hp_by_gear)
```

### ðŸ”§ Try It Yourself - Task 4 - Sorting Data
- 4.1 Sort the dataframe by mpg in descending order.
- 4.2 Sort the dataframe first by cyl and then by hp.

```python
# 4.1 Sort the dataframe by mpg in descending order.
df_sorted_mpg_desc = df.sort_values(by='mpg', ascending=False)
print("\nDataFrame sorted by mpg in descending order (first 5 rows):")
print(df_sorted_mpg_desc.head())

# 4.2 Sort the dataframe first by cyl and then by hp.
df_sorted_cyl_hp = df.sort_values(by=['cyl', 'hp'])
print("\nDataFrame sorted by cyl and then by hp (first 5 rows):")
print(df_sorted_cyl_hp.head())
```
